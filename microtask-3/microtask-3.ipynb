```python
# json library is used to handle json files, here, it is the data source 
# retrieved by the perceval module.
import json 

# to handle the time formats, like to determine 'created_at' of an issue or pr.
from datetime import datetime 

# importing pandas
# pandas is used for handling huge data using dataframe
import pandas as pd

# importing the plotting framework
import matplotlib.pyplot as plt
```

## Summarization Functions
Extracts the useful attribute from the raw data in the json file

```python
def summarize_issue(issue):
    """
    Get the contents of the issue/pr.
    
    This method gives, by taking the line data, 
    the summary of the issue/pr.
    
    :param item: line json data of the issue/pr
    :return: content of the line
    """
    repo = issue['origin']
    cdata = issue['data']
    summary ={
            'repo': repo,
            'hash': cdata['id'],
            'author': cdata['user']['login'],
            'created_date': datetime.strptime(cdata['created_at'],"%Y-%m-%dT%H:%M:%SZ"),
            'closed_date': datetime.strptime(cdata['closed_at'],"%Y-%m-%dT%H:%M:%SZ")
                                        if cdata['closed_at'] else None,
            'state':cdata['state'],
            'timestamp':datetime.utcfromtimestamp(int(issue['timestamp'])).strftime('%Y-%m-%d %H:%M:%S')
    }
    return summary
```

```python
def get_content(repo):
    """
    Get the contents of the project.
    
    This method gives, by taking the data retrived by perceval, 
    the content of the repository.
    
    :param repo: get the name of the repository
    
    :return: contents dataframe of the repository
    """
    # intializing the content dataframe with the required details
    content_df = pd.DataFrame(columns=['hash', 'author', 'created_date', 
                                       'state','closed_date','timestamp'])

    # contents is to store the details of each issue
    content = []
    
    # we need only issues here
    # loading the file into an object
    with open('../data/%s.json'%repo) as datasrc:
        # iterating through the lines od the data source
        for line in datasrc:
            # load the line in the json format so as to iterate to get the required results
            line = json.loads(line)
            # if it is a issue, get the details of issue
            if line['category'] == 'issue':
                # calling the summerize issue function
                summary = summarize_issue(line)
                # appending the data in the content list
                content.append(summary)
    # appending in the dataframe
    content_df = content_df.append(content,sort= False)
    # returning the dataframe
    return content_df
```

```python
# define the names of the repos which you want to analyze
repos = ['badgeyay','phimpme-android','susi_server','susi_android','open-event-server']

# iterating through the repositories
for repo in repos:
    # get the contents of the dataframe
    content_df = get_content(repo)
    # deriving the close time from the content dataframe
    close_time=(content_df['closed_date']-content_df['created_date']).astype('timedelta64[h]')
    # plotting the close time versus frequency in histograms
    close_time.plot.hist( bins=20,range=(0,200))
    # assign xlabel
    plt.xlabel('%s Close Issue Time (Hours)'%repo);
    # saving the plot
    plt.savefig('%s.png'% repo)
    # showing the plot
    plt.show()
```
