```python
import json
import datetime
import csv

import requests
import matplotlib.pyplot as plt

from collections import defaultdict
from prettytable import from_csv

from pprint import pprint
```

```python
class Repository_Summary:
    
    def __init__(self, path):
        """
        """
        self.repodata = defaultdict(list)
        
        self.pr_count={'open':0, 'rejected':0, 'merged':0}
        
        self.df = {
            # commit has the column >> hash, author, date
            'commit':pd.DataFrame(columns=['hash', 'author', 'author_date',
                                           'commit', 'created_date', 'files_no']),
            # issue has the column >> hash, author, date
            'issue':pd.DataFrame(columns=['hash', 'author', 'created_date',
                                         'updated_date', 'comments', 'state']),
            # pr has the column >> hash, author, date
            'pull_request':pd.DataFrame(columns=['hash', 'author', 'created_date', 
                                        'closed_date', 'merged_date', 'comments','commits',
                                        'additions', 'deletions', 'changed_files', 'state'])
        }
        
        contents = defaultdict(list)

        # to filter out commit, issue, pr details from the data source and store them seperately in dict.
        with open('%s'%path) as datasrc:
            for line in datasrc:
                line = json.loads(line)
                if line['category'] == 'commit':    
                    content = self.summary_commit(line) 
                elif line['category'] == 'issue':    
                    content = self.summary_issue(line)
                elif line['category'] == 'pull_request':    
                    content = self.summary_pr(line) 
                contents[line['category']].append(content)
        for item in contents:
            self.df[item]=self.df[item].append(contents[item],sort=False)
            
            
    def summary_commit(self,commit):
        repo = commit['origin']
        data = commit['data']
        summary ={
                'repo': repo,
                'hash': data['commit'],
                'author': data['Author'],
                'author_date': datetime.datetime.strptime(data['AuthorDate'],
                                                          "%a %b %d %H:%M:%S %Y %z"),
                'commit': data['Commit'],
                'created_date': datetime.datetime.strptime(data['CommitDate'],
                                                          "%a %b %d %H:%M:%S %Y %z"),
                'files_no': len(data['files'])
        }
        actions = 0
        for file in data['files']:
            if 'action' in file:
                actions += 1
        summary['files_action'] = actions
        summary['merge'] = 'Merge' in data
        return summary
    
    def summary_issue(self,issue):
        repo = issue['origin']
        data = issue['data']
        summary ={
                'repo': repo,
                'hash': data['id'],
                'author': data['user']['login'],
                'created_date': datetime.datetime.strptime(data['created_at'],"%Y-%m-%dT%H:%M:%SZ"),
                'updated_date': datetime.datetime.strptime(data['updated_at'],"%Y-%m-%dT%H:%M:%SZ")
                                            if data['updated_at'] else None,
                'comments': data['comments'],
                'state': data['state']
        }
        return summary
    
    def summary_pr(self,pr):
        repo = pr['origin']
        data = pr['data']
        summary ={
                'repo': repo,
                'hash': data['id'],
                'author': data['user']['login'],
                'created_date': datetime.datetime.strptime(data['created_at'],"%Y-%m-%dT%H:%M:%SZ"),
                'closed_date': datetime.datetime.strptime(data['closed_at'],"%Y-%m-%dT%H:%M:%SZ")
                                            if data['closed_at'] else None,
                'merged_date': datetime.datetime.strptime(data['merged_at'],"%Y-%m-%dT%H:%M:%SZ")
                                            if data['merged_at'] else None,

                'comments': data['comments'],
                'commits': data['commits'],
                'additions': data['additions'],
                'deletions': data['deletions'],

                'changed_files': data['changed_files'],
                'state': data['state']
        }  
        return summary
    
    def count_pr(self):
        repo_df=self.df
        count=self.pr_count
        for item in repo_df["pull_request"]:
            if(item['state']=="open"):
                count['open']+=1
            elif(item["state"]=="closed"):
                if (item["closed_date"] != None and item["merged_date"] == None):
                    count['rejected']+=1
                elif(item["closed_date"] != None and item["merged_date"] != None):
                    count["merged"]+=1
            self.pr_count=count

    
    def print_pr_count(self):
        print("Open PR: ",pr_count['open'])
        print("Total Closed: ", pr_count['rejected']+pr_count['merged'])
        print("\tClosed but not Merged: ", pr_count['rejected'])
        print("\tClosed and Merged: ", pr_count['merged'])

    def plot_open_close(self):
        patches, texts = plt.pie([self.pr_count['open'], 
                                  self.pr_count['rejected']+self.pr_count['merged']],
                         colors=['yellowgreen', 'gold'], shadow=True,
                         startangle=90)
        plt.legend(patches, ['Open', 'Closed'], loc="best")
        plt.axis('equal')
        plt.tight_layout()
        plt.show()
        
    def plot_status(self):
        patches, texts = plt.pie([self.pr_count['open'], 
                                  self.pr_count['rejected']+self.pr_count['merged']],
                                 colors=['yellowgreen', 'lightskyblue', 'lightcoral'],
                                 shadow=True, startangle=90)
        plt.legend(patches, ['Open', 'Closed but not Merged', 'Closed and Merged'], loc="best")
        plt.axis('equal')
        plt.tight_layout()
        plt.show()
```
